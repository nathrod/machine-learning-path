## Understanding how images are read by the computer

Color image --> Three-dimensional matrix --> height, width, and color depth (the color channels)

In Python --> matrix of matrices

*   Height (H): number of rows (vertical pixels).
*   Width (W): number of columns (horizontal pixels).
*   Color Channels (C): usually 3 (for Red, Green, Blue).

*HxWxC*

```python
#normalizing
pixels = np.array([
    [0, 128, 255],
    [64, 192, 0]
])

normalized_pixels = pixels / 255.0
print(normalized_pixels)
```

## Calculate the intensity of gray for each pixel.

Luminance formula: =0.299 + 0.587 +0.114

1. Iterate over each pixel of the image.
2. Apply the formula.
3. Replace the RGB value with 

```python
cat = Image.open("cat.png")
cat = np.asarray(cat, dtype=np.float32)/255
```

```python
plt.figure(figsize=(2,4))
plt.imshow(cat)
#imag = plt.imshow(img,aspect='auto')
plt.axis("off")
plt.show()
```

```python
size = cat.shape
size
```

```python
cat[0, 0, :]
```

```python
new_cat = np.zeros((size[0], size[1]), dtype=np.float32)
new_cat.shape
```

```python
for i in range(size[0]):
  for j in range(size[1]):
    new_cat[i, j] = cat[i, j, 0]*0.299 + cat[i, j, 1]*0.587 + cat[i, j, 2]*0.114
```

```python
new_cat = new_cat * 255
new_cat = new_cat.astype(np.uint8)
gray_image = Image.fromarray(new_cat)
```

```python
plt.figure(figsize=(2,4))
plt.imshow(new_cat,cmap='gray')
plt.axis("off")
plt.show()
```

```python
black_white_cat = np.zeros((new_cat.shape[0], new_cat.shape[1]), dtype=np.float32)
black_white_cat.shape
```

```python
new_cat
```

```python
for i in range(new_cat.shape[0]):
  for j in range(new_cat.shape[1]):
    if new_cat[i][j] < 127.5:
      black_white_cat[i][j] = 0
    elif new_cat[i][j] >= 127.5:
      black_white_cat[i][j] = 1
```

```python
black_white_cat
```

```python
plt.figure(figsize=(2,4))
plt.imshow(black_white_cat,cmap='gray')
plt.axis("off")
plt.show()
```
